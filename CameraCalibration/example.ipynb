## 角点检测

```python
import cv2

# 读取图像
image = cv2.imread('calib.png')
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# 棋盘格尺寸
pattern_size = (7, 10)

# 检测棋盘格角点
ret, corners = cv2.findChessboardCorners(gray, pattern_size)

if ret:
    # 在图像上绘制角点
    cv2.drawChessboardCorners(image, pattern_size, corners, ret)
    cv2.imshow('Chessboard Corners', image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
else:
    print("Failed to find chessboard corners.")
```

```python
print(corners)
print(corners.size)
print(corners.shape)
```

## 提取视频帧

```python
import cv2
videoCapture = cv2.VideoCapture("videoFrames.mp4")
success, image = videoCapture.read()
success
```

```python
import os
output_folder = "videoFrames"
if not os.path.exists(output_folder):
    os.makedirs(output_folder)
```

```python
count = 0

# 提取视频中的所有帧
while success:
    frame_filename = f"{output_folder}/frame_{count}.jpg"
    cv2.imwrite(frame_filename, image)  # 保存帧图像
    success, image = videoCapture.read()
    print(f"Frame {count} extracted successfully")
    count += 1

videoCapture.release()
```

```python
import os
import glob

"""
需要修改的三个点：
- 改名文件所在文件夹的位置：img_folder
- 改名文件的后缀名：file_ext
- 新文件名的格式：new_img_name
"""
# 指定修改的目录和文件后缀名
img_folder = "images"
file_ext = ".jpg"

images = glob.glob(f"{img_folder}/*{file_ext}")   # 获取以jpg为后缀的文件目录

# 遍历每个图像文件
for idx, img_path in enumerate(images):
    img_name, img_ext = os.path.splitext(os.path.basename(img_path))   # img_ext保存文件扩展名
    new_img_name = f"image00{idx + 1}{img_ext}"   # 定义新文件名格式
    new_img_path = os.path.join(img_folder, new_img_name)
    
    # 重命名图像文件
    os.rename(img_path, new_img_path)
    
    print(f"Renamed {img_path} to {new_img_path}")
```

## 单视角图像校正

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

"""
fx、fy:焦距
cx、cy:主点位置
"""
fx = 2.982248739396491e+03
fy = 2.990129977671481e+03
cx = 1.510075419524916e+03
cy = 1.600423574355771e+03



K = np.array([[fx, 0, cx],
              [0, fy, cy],
              [0, 0, 1]])

# 设置径向畸变系数（假设已知）
k1 = 0.201810138259160
k2 = -0.452868774440220

dist_coeffs = np.array([k1, k2, 0, 0, 0])

# 读取要校正的图像
# img = cv2.imread('images/image004.jpg')
img = plt.imread('images/image004.jpg')

# 对图像进行畸变校正
img_corrected = cv2.undistort(img, K, dist_coeffs)

# 显示校正后的图像
# cv2.imshow('Corrected Image', img_corrected)
# cv2.waitKey(0)
# cv2.destroyAllWindows()

fig, axes = plt.subplots(1, 2)
fig.set_size_inches(16, 10)
for ax in axes:
    ax.figure.set_size_inches(8, 5)
    ax.figure.set_dpi(150)
axes[0].imshow(img)
axes[0].set_title('Image 1')
axes[1].imshow(img_corrected)
axes[1].set_title('Image 2')

for ax in axes:
    ax.axis('off')

plt.show()
```
